## Task 1: Exploring and Visualizing the Iris Dataset

#### 1. Introduction & Problem Statement

The objective of this task is to explore and visualize the Iris dataset.
We will analyze dataset structure, summarize data, and create visualizations such as scatter plots, histograms, and box plots.

This task demonstrates skills in:

- Data inspection

- Exploratory Data Analysis (EDA)

- Data visualization

#### 2. Import Required Libraries

```python
# Data Handling
import pandas as pd

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
from sklearn.datasets import load_iris
```

Explanation:

- pandas → data manipulation

- matplotlib & seaborn → visualization

- load_iris → built-in dataset

#### 3. Load Dataset

```python
iris = load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

df.head()
```

#### 4. Dataset Structure

```python
print("Shape:", df.shape)
print("Columns:", df.columns)
print("\nSummary Statistics:")
print(df.describe())
```

#### 5. Scatter Plot

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='sepal length (cm)', 
                y='sepal width (cm)', 
                hue='species', 
                data=df)
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

#### 6. Histogram

```python
plt.figure(figsize=(6,4))
sns.histplot(df['petal length (cm)'], kde=True)
plt.title("Distribution of Petal Length")
plt.show()
```

#### 7. Box Plot

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='species', y='petal width (cm)', data=df)
plt.title("Petal Width by Species")
plt.show()
```

#### 8. Conclusion

- Iris dataset contains 150 samples and 4 numerical features.

- Petal features show clear separation among species.

- Visualization provides better understanding of feature distribution.
